# Feature extraction

```python
clean_copy = tit_clean.copy()

tit_clean.head()
```

```python
tit_clean.dtypes
```

```python
cat_columns = ['Name', 'Sex', 'Ticket ', 'Embarked']

cat_columns
```

```python
# Transforming categorical data to numerical data
from sklearn.preprocessing import LabelEncoder
```

```python
# label encode thte data
# instantiate the encoder
encoder = LabelEncoder()

# fit and transform the dataset
for column in cat_columns:
    tit_clean[column] = encoder.fit_transform(tit_clean[column])
```

```python
#encoded dataset
tit_clean.head()
```

```python
tit_clean['Embarked'].unique()
```

```python
mapping = {'Q':1, 'S':2, 'C':0}

tit_clean['Embarked'].replace(mapping, inplace=True)

tit_clean.head()
```

```python
#correlation dataset
corr_matrix = tit_clean.corr()

# plot heatmap of correlation
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='inferno')
```

```python
tit_clean.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)

tit_clean.head()
```

```python
# import the libraries
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix , classification_report
```

```python
x = tit_clean.drop('Survived', axis=1)
y = tit_clean['Survived']
```

```python
#split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify = y)
```

```python
# Start with dummy clcassifier to establish the baseline
dummy = DummyClassifier()
dummy.fit(x_train,y_train)
dummy_pred = dummy.predict(x_test)
```

```python
confusion_matrix(y_test, dummy_pred)
```

```python
print(classification_report(y_test, dummy_pred))
```

```python
# logistic regression
logr= LogisticRegression()
logr.fit(x_train, y_train)
logr_pred = logr.predict(x_test)
```

```python
# confussion matrix of logistic regression model
confusion_matrix(y_test, logr_pred)
```

```python
print(classification_report(y_test, logr_pred))
```

```python
# Decision Tree Model
dect= DecisionTreeClassifier()
dect.fit(x_train, y_train)
dect_pred = dect.predict(x_test)
```

```python
# confussion matrix of Decision Tree model
confusion_matrix(y_test, dect_pred)
```

```python
print(classification_report(y_test, dect_pred))
```

```python
# Random Forest Model
ranf= RandomForestClassifier()
ranf.fit(x_train, y_train)
ranf_pred = ranf.predict(x_test)
```

```python
# confussion matrix of Random Forest model
confusion_matrix(y_test, ranf_pred)
```

```python
print(classification_report(y_test, ranf_pred))
```
